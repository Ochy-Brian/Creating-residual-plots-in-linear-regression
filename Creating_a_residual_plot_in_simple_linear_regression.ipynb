Import packages and libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

Reading the csv file

```python
data = pd.read_csv('/content/headbrain3.csv')
data.head()
```

Fit simple linear regression model

```python
linear_model = ols('Brain_weight ~ Head_size', data=data).fit()
```

Display the model summary

```python
print(linear_model.summary())
```

modify figure size

```python
fig = plt.figure(figsize=(14, 8))
```

creating regression plots

```python
fig = sm.graphics.plot_regress_exog(linear_model,
                                    'Head_size',
                                    fig=fig)
fig
```

In conclusion we can see that the points are plotted randomly spread or scattered.points or residuals are scattered around the '0' line,there is no pattern, and points are not based on the side so there is no problem of heteroscedastisity.with the predictor variable 'Head_size' theres no heteroscedasticity
